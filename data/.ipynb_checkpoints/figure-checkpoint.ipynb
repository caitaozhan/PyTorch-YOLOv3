```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
import seaborn as sns
import numpy as np
```

```python
idx = 1
sample_per_label = 10
dataset = "61test"

fig, axs = plt.subplots(1, 3, figsize=(30, 10))

folder = int(idx / sample_per_label)
folder = format(folder, '06d')
matrix_name = str(idx % sample_per_label) + '.npy'
axs[0].axis('off')
img = np.load(f'/home/caitao/Project/dl-localization/data/{dataset}/{folder}/{matrix_name}')
axs[0].set_title('Input Sensing Data', fontsize=20)
axs[0].imshow(img, cmap='gray')

axs[1].axis('off')
idx = format(idx, '06d')
img = np.load(f'samples2/{idx}.npy')
axs[1].set_title('Image Translation', fontsize=20)
axs[1].imshow(img)

axs[2].axis('off')
img = mpimg.imread(f'../output/{idx}.png')
axs[2].imshow(img)
```
